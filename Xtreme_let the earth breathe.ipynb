```python
#Packages / Libraries

import os
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
%matplotlib inline

#I learned via youtube so I just put all the packages and libraries that I saw from the video reference that I watched.
```

```python
#Loading the Data

fossil = pd.read_csv('xtreme.csv')

fossil
```

```python
plt.figure(figsize=(15, 10))
plt.plot(fossil.Year, fossil['Asia Pacific'])
plt.plot(fossil.Year, fossil.Azerbaijan)
plt.plot(fossil.Year, fossil.Bangladesh)
plt.plot(fossil.Year, fossil.China)
plt.plot(fossil.Year, fossil.CIS)
plt.plot(fossil.Year, fossil.Cyprus)
plt.plot(fossil.Year, fossil['Hong Kong'])
plt.plot(fossil.Year, fossil.India)
plt.plot(fossil.Year, fossil.Indonesia)
plt.plot(fossil.Year, fossil.Iran)
plt.plot(fossil.Year, fossil.Iraq)
plt.plot(fossil.Year, fossil.Israel)
plt.plot(fossil.Year, fossil.Japan)
plt.plot(fossil.Year, fossil.Kazakhstan)
plt.plot(fossil.Year, fossil.Kuwait)
plt.plot(fossil.Year, fossil.Malaysia)
plt.plot(fossil.Year, fossil['Middle East'])
plt.plot(fossil.Year, fossil.Oman)
plt.plot(fossil.Year, fossil.Pakistan)
plt.plot(fossil.Year, fossil.Philippines)
plt.plot(fossil.Year, fossil.Qatar)
plt.plot(fossil.Year, fossil.Russia)
plt.plot(fossil.Year, fossil['Saudi Arabia'])
plt.plot(fossil.Year, fossil.Singapore)
plt.plot(fossil.Year, fossil['South Korea'])
plt.plot(fossil.Year, fossil['Sri Lanka'])
plt.plot(fossil.Year, fossil.Taiwan)
plt.plot(fossil.Year, fossil.Thailand)
plt.plot(fossil.Year, fossil.Turkey)
plt.plot(fossil.Year, fossil.Turkmenistan)
plt.plot(fossil.Year, fossil['United Arab Emirates'])
plt.plot(fossil.Year, fossil.Uzbekistan)
plt.plot(fossil.Year, fossil.Vietnam)





plt.show()
```

```python

a = fossil

a[["Argentina", "Brazil", "Chile", "Colombia", "Ecuador", "Peru", "Venezuela"]].plot(kind="bar")
plt.title("2014-2020 South America Fossil Fuel Consumption", fontsize = 25)
plt.legend(loc='lower center', bbox_to_anchor=(0.5, 1.20), ncol=7, fancybox=True, shadow=True)
plt.xticks(range(0,len(a)),('2014', '2015', '2016', '2017', '2018', '2019', '2020'))

plt.show()
```

```python
a = fossil

a[["Algeria", "Eastern Africa", "Egypt", "Middle Africa", "Morocco", "South Africa", "Western Africa"]].plot(kind="area")
plt.title("2014-2020 Africa Fossil Fuel Consumption", fontsize = 25)
plt.legend(loc='lower center', bbox_to_anchor=(0.5, 1.20), ncol=7, fancybox=True, shadow=True)
plt.xticks(range(0,len(a)),('2014', '2015', '2016', '2017', '2018', '2019', '2020'))

plt.show()

```

```python
ax = sns.lineplot(x = 'Year', y = 'Australia', data = fossil)
```

```python


fig, axis = plt.subplots(1,2, figsize =(20,7))

ax = sns.lineplot(x = 'Year', y = 'Australia', data = fossil, ax = axis[0])
ax = sns.lineplot(x = 'Year', y = 'New Zealand', data = fossil, ax = axis[1])

fig.suptitle('2014-2020 Australia and Oceania Fossil Fuel Consumption', fontsize=30)
```

```python
fig, axis = plt.subplots(2,2, figsize =(20,7))

ax = sns.barplot(x = 'Year', y = 'Canada', data = fossil, ax = axis[0,0])
ax = sns.barplot(x = 'Year', y = 'Mexico', data = fossil, ax = axis[0,1])
ax = sns.barplot(x = 'Year', y = 'Trinidad and Tobago', data = fossil, ax = axis[1,0])
ax = sns.barplot(x = 'Year', y = 'United States', data = fossil, ax = axis[1,1])

fig.suptitle('2014-2020 North America Fossil Fuel Consumption', fontsize=30)






  
```
